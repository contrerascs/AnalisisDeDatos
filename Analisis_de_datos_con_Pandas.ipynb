## Analisis de edades

```python
data['Gender'].value_counts()
```

```python
data['Age'].mean()
```

```python
data['Age'].std()
```

```python
data.loc[data['Age'] <= 40, 'Age_group'] = 'Adult'
data.loc[data['Age'] >= 41, 'Age_group'] = 'Older Adult'
```

```python
data.head()
```

```python
data.tail()
```

```python
data['Age_group'].value_counts()
```

```python
data.groupby('Gender')['Age'].agg({'mean','std'})
```

## Analisis de las ocupaciones

```python
data.groupby('Occupation')['Gender'].value_counts()
```

```python
print(data['Sleep Duration'].min(), data['Sleep Duration'].max())
```

```python
data.groupby('Gender')[['Sleep Duration', 'Quality of Sleep']].mean()
```

```python
#CALIDAD Y DURACION DEL SUEÑO EN DISTINTAS OCUPACIONES
occupation_sleep = data.groupby('Occupation')[['Sleep Duration', 'Quality of Sleep']].mean()
#ORDENAMIENTO DE DATOS PARA UNA MEJOR VISUALIZACION
occupation_sleep.sort_values(by='Quality of Sleep')
```

```python
#EL INDICE DE MASA CORPORAL POR OCUPACIONES
occupation_sleep = data.groupby('Occupation')['BMI Category'].value_counts()
#ORDENAMIENTO DE DATOS PARA UNA MEJOR VISUALIZACION
occupation_sleep.sort_values()
```

```python
occupation_steps = data.groupby('Occupation')[['Daily Steps','Physical Activity Level','Sleep Duration','Quality of Sleep']].mean()
occupation_steps.sort_values(by='Quality of Sleep')
```
