```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('/content/top_rated_9000_movies_on_TMDB.csv', index_col='id')
data.tail()
```

```python
data.info()
```

```python
data['Genres'].value_counts()
```

```python
language_counts = data['original_language'].value_counts()
language_percentages = data['original_language'].value_counts(normalize=True)*100
print(language_percentages.head())
```

```python
# Configurar el estilo de los gráficos de Seaborn
sns.set(style='whitegrid')

# Crear un histograma de la columna 'rating'
plt.figure(figsize=(10, 6))
sns.histplot(data['vote_average'], bins=20, kde=True)

# Añadir títulos y etiquetas
plt.title('Distribución de Calificaciones de las Películas')
plt.xlabel('Calificación')
plt.ylabel('Número de Películas')

# Mostrar el gráfico
plt.show()
```
